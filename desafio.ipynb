# Projeto para redução de dimensionalidade em imagens.



```python
import cv2
from google.colab.patches import cv2_imshow
```

Baixando imagem

```python
%%time
!curl -o cachorro.png https://img.olx.com.br/thumbs256x256/78/785317604978580.jpg
import cv2
img = cv2.imread('cachorro.png', cv2.IMREAD_UNCHANGED)
cv2_imshow(img)
```

#1 Redimensionando o arquivo

```python
print("Dimensões do arquivo", img.shape)
```

```python
resized = cv2.resize(img, (255,255), interpolation = cv2.INTER_AREA)
```

```python
%%time
cv2_imshow(resized)
```

# 2 Tons de cinza

```python
gray_image = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
```

```python
%%time
cv2_imshow(gray_image)
```

# 3 Preto e branco

```python
(thresh, blackAndWhiteImage) = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
```

```python
cv2_imshow(blackAndWhiteImage)
```

# 4 Função para conversão

```python
def binarizar_imagem(img_path):
  original = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)
  resized = cv2.resize(original, (255,255), interpolation = cv2.INTER_AREA)
  gray_image = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
  (thresh, blackAndWhiteImage) = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

  return original, gray_image, blackAndWhiteImage
```

```python
original, escala_cinza, preto_branco = binarizar_imagem('cachorro.png')
```

```python
cv2_imshow(original)
cv2_imshow(escala_cinza)
cv2_imshow(preto_branco)
```
